```python
from matplotlib import pyplot as plt

print(plt.boxplot.__doc__)
```

```python
plt.boxplot([[1,2,3],[2,1,4]],
            notch=False,
            vert=False,
            usermedians=[2.1,2.1],
            widths=[0.1,0.2],
            patch_artist=True,
            labels=['first', 'second'],
            manage_ticks=False)

plt.show()
```
