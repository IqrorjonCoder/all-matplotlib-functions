```python
from matplotlib import pyplot as plt

print(plt.barh.__doc__)
```

```python
x = plt.barh(y=[1,3,5],
             width=[2,1,3],
             height=1,
             left=[0.1, 0.2, 0.1],
             align='center',
             color=['red', 'yellow', 'green'],
             edgecolor='blue',
             linewidth=2,
             tick_label=['one', 'two', 'three'])

plt.bar_label(x, ['first', 'second', 'third'], label_type='center', color='blue')

plt.show()
```
