```python
from matplotlib import pyplot as plt

print(plt.axline.__doc__)
```

```python
plt.plot([1,2,3,2,1,3],
         [2,3,4,2,1,2])

plt.axline((1.5,1.5), (1.5,3.5), linewidth=2, color='red', marker='o',
           markeredgecolor='blue',
           markerfacecolor='green',
           markeredgewidth=5,
           markersize=10,
           visible=True,
           zorder=10)

plt.show()
```
