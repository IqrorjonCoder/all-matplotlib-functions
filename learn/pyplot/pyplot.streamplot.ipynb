```python
from matplotlib import pyplot as plt
import numpy as np

print(plt.streamplot.__doc__)
```

```python
x = np.arange(0, 10)
y = np.arange(0, 10)

X, Y = np.meshgrid(x, y)

u = np.ones((10, 10))

v = np.zeros((10, 10))

fig = plt.figure(figsize = (12, 7))

plt.streamplot(X, Y, u, v, density = 0.5)

plt.show()
```
