```python
from matplotlib import pyplot as plt

print(plt.bar.__doc__)
```

```python
plt.bar(x=[1,2,3],
        height=[2,1,3],
        width=0.5,
        bottom=0.1,
        align='center',
        color=['red', 'blue', 'green'],
        edgecolor='yellow',
        linewidth=2,
        tick_label=['simple1','simple2','simple3'],
        log=False)

plt.show()
```
