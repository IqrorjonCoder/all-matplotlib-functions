```python
from matplotlib import pyplot as plt
import numpy as np

print(plt.close.__doc__)
```

```python
fig, ax = plt.subplots(2,1)
fig2, ax2 = plt.subplots(2,1)

ax[0].plot(np.random.random(30))
ax[1].plot(np.random.random(30))

ax2[0].plot(np.random.random(30))
ax2[1].plot(np.random.random(30))

fig.suptitle('first')
fig2.suptitle('second')

plt.close(1)

plt.show()
```
