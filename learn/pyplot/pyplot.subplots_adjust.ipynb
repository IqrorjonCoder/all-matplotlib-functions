```python
from matplotlib import pyplot as plt

print(plt.subplots_adjust.__doc__)
```

```python
fig, ax = plt.subplots()

ax.plot([2 , 4 , 6 , 8 , 10],
        [5 , 10 , 15 , 20 , 25])

fig.subplots_adjust(right=2,
                    left=0,
                    top=2,
                    bottom=0)

plt.show()
```
