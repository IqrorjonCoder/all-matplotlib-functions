```python
r = plt.imread("/home/iqrorjoon/PycharmProjects/all matplotlib functions/base/simple_image.jpg")
im = plt.imshow(r)
plt.colorbar(
    location='left',
    orientation='vertical',
    fraction=0.1,
    shrink=0.7,
    aspect=15,
    pad=0.2,
    anchor=(2.0, 0.5),
    extend='max',
    drawedges=True, #False,
    label='simple colorbar'
)

plt.show()
```
