```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 2, 1],
        [2, 3, 2, 1, 2])


def button_press_event(event):
    f = [i for i in dir(event) if i[0] != '_']
    print(event.button)
    if str(event.button) == "MouseButton.LEFT":
        cid = event.canvas.mpl_connect('button_press_event', button_press_event)
        event.canvas.mpl_disconnect(cid)
        print("disconnected !!!")

plt.connect('button_press_event', button_press_event)

plt.show()
```
