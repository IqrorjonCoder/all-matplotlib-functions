```python
from matplotlib import pyplot as plt

print(plt.axhspan.__doc__)

plt.plot([1,2,3,1,2],
         [2,3,1,2,3])

plt.axhspan(2.5, 2.75, facecolor='green', alpha=0.9)

plt.show()
```
