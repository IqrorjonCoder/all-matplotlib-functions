```python
from matplotlib import pyplot as plt

print(plt.fill_betweenx.__doc__)
```

```python
fig, ax = plt.subplots(2,1)

ax[0].fill_betweenx([1,2,1,3,2],
                   [2,3,1,2,3],
                   # [0,0,1]
                   # step='pre'
                   color='red'
                   )
ax[1].plot([1,2,1,3,2],
           [2,3,1,2,3])

plt.show()
```
