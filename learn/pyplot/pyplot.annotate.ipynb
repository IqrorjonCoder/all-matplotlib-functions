```python
import numpy as np
from matplotlib import pyplot as plt

x = np.array([1,2,3,3,2,2])
y = np.array([2,3,5,7,3,10])
plt.plot(x,y)

plt.annotate('simplee', xy=(2.0,8.0), style='italic',xytext=(1.1,9.1),
             arrowprops={
                 'arrowstyle': 'fancy',
                 'mutation_scale': 20.2,
             })

plt.show()

print(plt.annotate.__doc__)
from matplotlib.patches import FancyArrowPatch
```
