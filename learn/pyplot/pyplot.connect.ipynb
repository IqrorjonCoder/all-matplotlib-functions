```python
fig, ax = plt.subplots()
ax.plot([1,2,3,2], [3,2,1,2])

mouse = {
    1: 'left',
    2: 'middle',
    3: 'right',
    8: 'back',
    9: 'forward',
}


def on_click(event):
    print(f"you pressed {mouse[event.button]} mouse ||| ({event.xdata}, {event.ydata})")


plt.connect('button_press_event', on_click)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 2, 1],
        [2, 3, 2, 1, 2])


def release_event(event):
    f = [i for i in dir(event) if i[0] != '_']
    print("button: ", event.button)
    print("canvas: ", event.canvas)
    print("dblclick: ", event.dblclick)
    print("guiEvent: ", event.guiEvent)
    print("inaxes: ", event.inaxes)
    print("key: ", event.key)
    print("lastevent: ", event.lastevent)
    print("name: ", event.name)
    print("step: ", event.step)
    print("x, event.xdata: ", event.x, event.xdata)
    print("y, event.ydata: ", event.y, event.ydata)


plt.connect('button_release_event', release_event)

plt.show()
```

```python

fig, ax = plt.subplots()

ax.plot([1, 2, 3, 2, 1],
        [2, 3, 2, 1, 2])


def draw_event(event):
    f = [i for i in dir(event) if i[0] != '_']
    print("canvas: ", event.canvas)
    print("guiEvent: ", event.guiEvent)
    print("name: ", event.name)
    print("renderer: ", event.renderer)
    print(event.renderer.height)
    print(event.renderer.width)


plt.connect('draw_event', draw_event)

plt.show()
```
