```python
import numpy as np

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', s='d', data=data)
plt.xlabel('x - axis', labelpad=1)
plt.ylabel('Y - axis')
plt.cool()
plt.title('Cool Colourmap : plt.cool()')
plt.colorbar()
plt.show()
```
